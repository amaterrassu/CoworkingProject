```python
def categorize_income(row):
    if 'жиль' in row:
        return 'Покупка жилья'
    elif 'автомоб' in row:
        return 'Операция с автомобилем'
    elif 'образован' in row:
        return 'Образование'
    elif 'свадьб' in row:
        return 'Проведение свадьбы'
    elif 'недвижимос' in row:
        return 'Недвижимость'
```

```python
df['category_income'] = df['purpose'].apply(categorize_income)
df['category_income'].head(30)
```

```python
def categorize_total_income(digit):
    if digit > 0 and digit < 40 * 10 ** 3:
        return 'D'
    elif digit > 40 * 10 ** 3 and digit < 100 * 10 ** 3:
        return 'C'
    elif digit > 100 * 10 ** 3 and digit < 200 * 10 ** 3:
        return 'B'
    elif digit > 200 * 10 ** 3:
        return 'A'
```

```python
df['categorize_income'] = df['total_income'].apply(categorize_total_income)
df.head()
```
